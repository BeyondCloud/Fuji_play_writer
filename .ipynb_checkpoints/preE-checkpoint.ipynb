```python
import copy as cp
with open('./raw/三隻小豬.txt','r', encoding = 'utf8') as f:
    txt = f.read()

txt_split = txt.split('\n')
txt_tmp = []
for t in txt_split:
    if t != '':
        txt_tmp.append(t)
txt_split = txt_tmp
filter_word = ['卻','了']
for t in txt_split:
    print(t,'\n')
    _t = cp.copy(t)
    for fw in filter_word:
        _t = _t.replace(fw,'') 
    print('>',_t+'\n')
```
