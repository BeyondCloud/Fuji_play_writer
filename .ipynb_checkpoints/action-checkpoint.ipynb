```python
def FindCertainChar(s, number):
    
    if number == 0:
        return s
    else:
        for i in range(number):
            loc = s.index('「', 0, len(s)) 
            s = s.replace(s[loc],'」',1)
            s = s.replace(s[loc+1], '&',1)
        return s
```

```python
'''
Basic idea:
1.Delete sentences which contain '「' and '」'
2.'&' is used for decide if an element should be deleted after the whole sentence is splited by '」'
3.Split again by '。'
4.'說'and'想' are seen as keywords which make a sentence fail to be an action
5. The rest of elements in list are concatenated to be the final result

'''
with open('./raw/三隻小豬.txt','r', encoding = 'utf8') as f:
    txt = f.read()

txt_split = txt.split('\n')
txt_tmp = []
for t in txt_split:
    if t != '':
        txt_tmp.append(t)
txt_split = txt_tmp

for t in txt_split:
    print('原句->', t)
    tmp = t
    times = tmp.count('「', 0 , len(tmp))
    s_final = FindCertainChar(tmp,times)
    s_final = s_final.split('」')
    for item in s_final:
        if '&' in item:
            s_final.remove(item)
    semi_str = ''
    for item in s_final:
        semi_str += item
    
    semi_str = semi_str.split('。',-1)
    for item in semi_str:
        if '說' in item or '想' in item:
            semi_str.remove(item)
    final_str = ''
    for item in semi_str:
        final_str += item
    
    final_str = final_str.replace('&','')
    print('結果->',final_str, '\n')
```
